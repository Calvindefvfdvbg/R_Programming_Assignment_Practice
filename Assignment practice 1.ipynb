```r
# Load necessary libraries
library(dplyr)
library(ggplot2)
library(lubridate)
```

```r
# Create a data frame
data <- data.frame(
  id = c(1, 2, 3, 4, 5, 6, 7),
  name = c("A", "B", "C", "D", "E", "F", "G"),
  department = c("HR", "Finance", "Mathematics", "Science", "English", "Chemistry", "Biology"),
  salary = c(60754, 59640, 69040, 65043, 59943, 65000, 69000),
  projects = c(4, 2, 8, 5, 2, 5, 7)
)

# Write the data frame to a CSV file
write.csv(data, "loan_data.csv", row.names = FALSE)
```
